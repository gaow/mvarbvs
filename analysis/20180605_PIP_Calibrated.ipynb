# Calibration of SNP level PIP

Calibrated PIP means that for example for SNPs having PIP = 0.5, indeed half of them are associated.

Here for susie and DAP we group all PIP across all simulations into 10 bins according to reported PIP. Then compute the proportion of truly associated SNPs in each bin. We align this proportion to the average PIP for each bin for calibrated SNP-level posterior probability.
