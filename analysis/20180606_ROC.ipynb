```sos
d1 = readRDS('susie_comparison/ROC_0605_estvar_true.rds')
d2 = readRDS('susie_comparison/ROC_0605_estvar_false.rds')
colors = c('#348ABD', '#7A68A6', '#A60628', '#467821', '#FF0000', '#188487', '#E2A233',
                  '#A9A9A9', '#000000', '#FF00FF', '#FFD700', '#ADFF2F', '#00FFFF')
```

```sos
plot(d1$susie$true_disc ~ d1$susie$false_disc, t="l", col=colors[1], ylab = "True discoveries", xlab ="False discoveries")
lines(d2$susie$true_disc ~ d2$susie$false_disc,col=colors[2])
lines(d1$dap$true_disc ~ d1$dap$false_disc,col=colors[3])
legend("bottomright", legend=c("susie (est_var)", "susie (var(Y)))", "DAP"),
       col=colors[1:3], lty=c(1,1,1), cex=0.8)
```
