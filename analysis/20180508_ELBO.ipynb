# M&M ELBO

Using `DSC/finemapping.ipynb` benchmark, I evaluate for situation with different effects the ELBO updates, under [current implementation](https://github.com/gaow/mvarbvs/tree/ed8cb6bc52969211e8cd045d7b7a1ba79ba75882), which I believe is still incorrect (the way single effect posterior is computed is problematic).

## J = 1000

```sos
cd ~/GIT/github/mvarbvs/DSC
dsc mnm.dsc --target debug_mnm_2
```

```sos
fitted_again = readRDS('benchmark/fit_mnm_debug/liter_data_1_original_Y_1_get_sumstats_1_init_mnm_1_fit_mnm_debug_1.rds')$fitted
plot(sapply(1:20, function(i) fitted_again[[i]]$elbo), ylab = 'ELBO', xlab = 'n_iter')
```
